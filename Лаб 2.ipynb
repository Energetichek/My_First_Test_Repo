```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

df = pd.read_excel('Solar_Taj.xlsx')

df['Date'] = pd.to_datetime(df['Date'])

df_first_month = df[df['Date'].dt.month == 1]

df_daily = df_first_month.resample('D', on='Date').agg(['mean', 'max'])

fig, ax = plt.subplots()
ax.plot(df_daily.index, df_daily['Insolation']['mean'], label='Average')
ax.plot(df_daily.index, df_daily['Insolation']['max'], label='Maximum')
ax.set_xlabel('Date')
ax.set_ylabel('Insolation (W/m^2)')
ax.set_title('Solar Insolation for First Month')
ax.legend()
plt.show()
```
